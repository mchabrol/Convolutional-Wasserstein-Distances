# **Texture mixing with regularization**

How to use this notebook? 

1. Example of our regularization method with fixed rho. 
2. Examples of our regularization method with different rho and comparison with first order method. 

```python
#use this cell to be at the root of the project (if needed)
import os
print(os.getcwd())
os.chdir('../')
print(os.getcwd())
```

```python
from wasserstein.texture_mixing_new import compute_texture_mixing
from wasserstein.texture_mixing_with_reg import compute_texture_mixing_with_reg
from utils.image_treatment import preprocess_image
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
### Load data + preprocess

image_path1 = 'data/zebrure.png'
image_path2 = 'data/rayures.png'

image1 = preprocess_image(image_path1, new_size = (100,100))
image2 = preprocess_image(image_path2, new_size = (100,100))
textures = [image1, image2]
rho = [0.5, 0.5]
```

```python
### Compute both texture mixing: without and with reg

synthesis_without_reg = compute_texture_mixing(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
synthesis_with_reg = compute_texture_mixing_with_reg(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
```

```python
### Plots

plt.figure(figsize=(10,4))
plt.subplot(1,4,1)
plt.imshow(textures[0])
plt.title('Original $f^1$', fontsize=10)
plt.axis('off')
plt.subplot(1,4,2)
plt.imshow(synthesis_with_reg)
plt.title(r'$\rho = {}$, with regularization'.format(rho[0]), fontsize=10)
plt.axis('off')
plt.subplot(1,4,3)
plt.imshow(synthesis_without_reg)
plt.title(r'$\rho = {}$, without regularization'.format(rho[0]), fontsize=10)
plt.axis('off')
plt.subplot(1,4,4)
plt.imshow(textures[1])
plt.title('Original $f^2$', fontsize=10)
plt.tight_layout()
plt.axis('off')
```

```python
rho = [0.2, 0.8]
synthesis_without_reg_1 = compute_texture_mixing(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
synthesis_with_reg_1 = compute_texture_mixing_with_reg(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
rho = [0.5, 0.5]
synthesis_without_reg_2 = compute_texture_mixing(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
synthesis_with_reg_2 = compute_texture_mixing_with_reg(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
rho = [0.2, 0.8]
synthesis_without_reg_3 = compute_texture_mixing(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
synthesis_with_reg_3 = compute_texture_mixing_with_reg(textures, rho, num_scales = 4, num_orientations = 4, n_iter = 100)
```

```python
### Examples with different rhos

plt.figure(figsize=(13,6))
plt.subplot(2,5,1)
plt.imshow(textures[0])
plt.ylabel('Without \n regularization', fontsize=12)  
plt.axis('on')  
plt.xticks([])  
plt.yticks([])
plt.title('Original $f^1$', fontsize=12)

plt.subplot(2,5,2)
plt.imshow(synthesis_without_reg_1)
plt.title(r'$\rho = {}$'.format(0.2), fontsize=12)
plt.axis('off')
plt.subplot(2,5,3)
plt.imshow(synthesis_without_reg_2)
plt.title(r'$\rho = {}$'.format(0.5), fontsize=12)
plt.axis('off')
plt.subplot(2,5,4)
plt.imshow(synthesis_without_reg_3)
plt.title(r'$\rho = {}$'.format(0.8), fontsize=12)
plt.axis('off')
plt.subplot(2,5,5)
plt.imshow(textures[1])
plt.title('Original $f^2$', fontsize=12)
plt.axis('off')

plt.subplot(2,5,6)
plt.imshow(textures[0])
plt.title('Original $f^1$', fontsize=12)
plt.ylabel('With \n regularization', fontsize=12)  
plt.axis('on')  
plt.xticks([])  
plt.yticks([])
plt.subplot(2,5,7)
plt.imshow(synthesis_with_reg_1)
plt.title(r'$\rho = {}$'.format(0.2), fontsize=12)
plt.axis('off')
plt.subplot(2,5,8)
plt.imshow(synthesis_with_reg_2)
plt.title(r'$\rho = {}$'.format(0.5), fontsize=12)
plt.axis('off')
plt.subplot(2,5,9)
plt.imshow(synthesis_with_reg_3) #synthesis_with_reg_1
plt.title(r'$\rho = {}$'.format(0.8), fontsize=12)
plt.axis('off')
plt.subplot(2,5,10)
plt.imshow(textures[1])
plt.title('Original $f^2$', fontsize=12)
plt.axis('off')
plt.tight_layout()
```
